```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"C:\Users\cashk\OneDrive\Desktop\Projects\Customer Churn ML\Churn_predictor_ML\data\telco_comm_churn.csv")
pd.set_option('display.max_columns', None)
df.head()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.describe()
```

### Hanlding missing values

```python
df.isnull().sum()
```

```python
df.fillna({'TotalCharges': df['TotalCharges'].median()}, inplace=True)
```

```python
df.isnull().sum().sum()
```

### Feature Engineering

##### Creating New Features

```python
df['tenure'].min(), df['tenure'].max()
```

```python
df['tenure_group']= pd.cut(df['tenure'], bins=5, labels=['0', '1', '2', '3', '4'], include_lowest=True)
df['tenure_group']
```

```python
df['tenure_group'].info()
```

```python
df['tenure_group'] = df['tenure_group'].astype(float).astype(int)
df['tenure_group'].info()
```

```python
df.head()
```

```python
df['avg_monthly_per_tenure'] = df['TotalCharges'] / (df['tenure'] + 1)
```

```python
df.columns
```

### Encoding

```python
df = df.drop('customerID', axis = 1)
```

```python
df.head()
```

```python
binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling']
for col in binary_cols:
    df[col] = df[col].map({'Yes':1, 'No':0})
    
df.head()
```

```python
df['Churn'] = df['Churn'].map({'Yes':1,'No':0})
```

```python
df['gender'] = df['gender'].map({'Male':1,'Female':0})
```

```python
df.info()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
categorical_cols
```

```python
for col in categorical_cols:
    le = LabelEncoder()
    
    df[col] = le.fit_transform(df[col].astype(str))
```
